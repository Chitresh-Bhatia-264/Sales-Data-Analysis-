```python
#Polynomial Curve Fitting 
monthly_sales=monthly_sales.reset_index()
monthly_sales['Data Ordinal']=monthly_sales['Date'].map(pd.Timestamp.toordinal)
```

```python
x=monthly_sales['Data Ordinal']
y=monthly_sales['Sales Amount']
```

```python
import numpy as np
```

```python
coeffs=np.polyfit(x,y,deg=2)
poly=np.poly1d(coeffs)
```

```python
future_dates = pd.date_range(start=monthly_sales['Date'].max() , periods=4, freq='M')
future_ordinals = future_dates.map(pd.Timestamp.toordinal)
```

```python
forecasted_sales=poly(future_ordinals)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales['Date'], y, label='Actual Sales')
plt.plot(future_dates, forecasted_sales, label='Forecasted Sales', linestyle='--', color='red')
plt.title("Polynomial Forecasting of Monthly Sales")
plt.xlabel("Date")
plt.ylabel("Sales Amount")
plt.legend()
plt.show()
```
